### Using the Object Oriented Method

```python
heart_disease.head()
```

```python
over_50 = heart_disease [heart_disease ["age"] > 50]
over_50.head()
```

```python
## pyplot method

over_50.plot (kind = "scatter",
             x = "age",
             y = "chol",
             c = "target");
```

```python
## oo method

fig, ax =plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',
            x= 'age',
            y = 'chol',
             c= 'target',
            ax=ax);
```

```python
## OOP from scratch

fig, ax = plt.subplots(figsize = (10,6))

scatter = ax.scatter(x = over_50 ["age"],
                    y = over_50 ["chol"],
                    c = over_50 ["target"])

#customize the plot

ax.set(title = "Heart Disease and Cholestrol Level",
      xlabel = "Age",
      ylabel = "Cholestrol Level");

## add a legend

ax.legend(*scatter.legend_elements(), title = "target");

## add a horizontal line

ax.axhline(over_50["chol"].mean(),
          linestyle = '--');
```

```python
# subplots of age, chol

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10));


# add data to ax0

scatter = ax0.scatter(x = over_50["age"],
                     y = over_50 ["chol"],
                     c = over_50 ["target"]);


# Customize ax0

ax0.set(title="Heart Disease and Cholestrol Level",
       xlabel = "Age",
       ylabel = "Cholestrol Level");

# add a legend to ax0

ax0.legend(*scatter. legend_elements(), title = "target")


# adda mean line

ax0.axhline(over_50["chol"].mean(),
           linestyle ="--");


# add data to ax1

scatter = ax1.scatter(x = over_50["age"],
                     y = over_50 ["thalach"],
                     c = over_50 ["target"]);


# Customize ax1

ax1.set(title="Heart Disease and Cholestrol Level",
       xlabel = "Age",
       ylabel = "thalach");

# add a legend to ax0

ax1.legend(*scatter. legend_elements(), title = "target")


# adda mean line

ax1.axhline(over_50["thalach"].mean(),
           linestyle ="--");
```

```python
# subplots of age, chol

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10));


# add data to ax0

scatter = ax0.scatter(x = over_50["age"],
                     y = over_50 ["chol"],
                     c = over_50 ["target"],
                     cmap="winter");


# Customize ax0

ax0.set(title="Heart Disease and Cholestrol Level",
       xlabel = "Age",
       ylabel = "Cholestrol Level");

# change the axis limits
ax0.set_xlim([50,80])


# add a legend to ax0
ax0.legend(*scatter. legend_elements(), title = "target")


# adda mean line

ax0.axhline(over_50["chol"].mean(),
           linestyle ="--");


# add data to ax1

scatter = ax1.scatter(x = over_50["age"],
                     y = over_50 ["thalach"],
                     c = over_50 ["target"],
                     cmap="winter");


# Customize ax1

ax1.set(title="Heart Disease and Cholestrol Level",
       xlabel = "Age",
       ylabel = "thalach");

# add a legend to ax0
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
ax1.legend(*scatter. legend_elements(), title = "target")


# adda mean line

ax1.axhline(over_50["thalach"].mean(),
           linestyle ="--");
```
